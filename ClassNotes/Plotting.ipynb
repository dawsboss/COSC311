```python
x = [i for i in range(1,11)]
y = [2*i + 3 for i in x]
print(x)
print(y)
```

```python
# this uses the first argument as the values for the x-axis
# uses the second argument as the values for the y-axis
# i.e. prints the first list versus the second list
plt.plot(x,y)
```

```python
plt.plot(
    [1,2,1.8,2,1,1], 
    [1,1,1.5,2,2,1],
    # pyplot has tons of named arguments to try using:
    color = 'green',
    linestyle = 'dotted',
    marker = '+',
    label = 'first line'
)
plt.plot(
    [1.3,2.4,1.8,2.1,1.0,1], 
    [1.2,1.7,1.5,2.1,2.3,1],
    # pyplot has tons of named arguments to try using:
    color = 'red',
    linestyle = 'solid',
    marker = '*',
    label = 'second line'
)
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Here is a cool plot of some wacky shapes')
plt.xticks([])
```

```python
plt.scatter(
    [1,2,1.8,2,1,1], 
    [1,1,1.5,2,2,1]
)
```
